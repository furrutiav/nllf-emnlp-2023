```python
import os
import openai
import time
import pandas as pd
```

```python
openai.organization = ""#"X"
openai.api_key = "X"
openai.Model.list()
```

```python
def chatgpt(text, temp=0.7, max_t=5):
    try: 
        output = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                  {"role": "system", "content": text},
            ],
            temperature=temp,
            max_tokens=max_t,
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0
        )
        openai.Completion.create(
            model="text-davinci-003",
            prompt=text,

        )
    except:
        print("Thinking...")
        time.sleep(1)
        return chatgpt(text, temp=temp, max_t=max_t)
    return output
```

```python
template = lambda q, a, v: f"""Pregunta:"{q}". Respuesta:"{a}". {v}"""
q = "Agustín desea guardar unos dados de 1 cm cúbico en una caja que mide 5 cm de largo, 2 cm de ancho y 2 cm de alto. Su hermano dice que caben en la caja 15 dados.¿Es correcto lo que señala el hermano de Agustín? Explica con tus palabras."
a = "estamal por caben 20 dados"
v = "¿La respuesta ofrece argumentos o ejemplos para respaldar su afirmación?"
prompt = template(q, a, v)
prompt
```

```python
%%time
chatgpt(prompt, temp=0.7, max_t=5)["choices"][0]["message"]["content"]
```

```python
train_sample = pd.read_excel("data/train_sample_task_C1.xlsx", index_col=0)
train_sample
```

```python
%%time
v = "¿La respuesta brinda una respuesta directa y relacionada con la pregunta planteada?"
responses = []
for i, ix in enumerate(train_sample.index):
    q, a, label = train_sample.loc[ix]["Q A label".split()]
    a = a.replace("\n", " ")
    a = " ".join(a.split(" ")).strip()
    o = chatgpt(template(q, a, v), temp=0.7, max_t=5)["choices"][0]["message"]["content"]
    o = o.replace("\n", " ").strip()
    responses.append([ix, o])
    print(100*(i+1)/train_sample.shape[0], ix, o, label)
```

```python
train_v1 = train_sample.copy()
train_v1["chatgpt"] = [x[1] for x in responses]
train_v1
```

```python
train_v1.to_excel("chatgpt_responses_V1.xlsx")
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(train_v1["label"], train_v1["chatgpt"].apply(lambda x: int("No" in x)), digits=4))
```
