```python
import os
import openai
import time
import pandas as pd
```

```python
openai.organization = ""#"X"
openai.api_key = "X"
openai.Model.list()
```

```python
def chatgpt(text, temp=0.7, max_t=5):
    try: 
        output = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                  {"role": "system", "content": text},
            ],
            temperature=temp,
            max_tokens=max_t,
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0
        )
        openai.Completion.create(
            model="text-davinci-003",
            prompt=text,

        )
    except:
        print("Thinking...")
        time.sleep(1)
        return chatgpt(text, temp=temp, max_t=max_t)
    return output
```

```python
template = lambda q, a, v: f"""Pregunta:"{q}". Respuesta:"{a}". {v}"""
q = "AgustÃ­n desea guardar unos dados de 1 cm cÃºbico en una caja que mide 5 cm de largo, 2 cm de ancho y 2 cm de alto. Su hermano dice que caben en la caja 15 dados.Â¿Es correcto lo que seÃ±ala el hermano de AgustÃ­n? Explica con tus palabras."
a = "estamal por caben 20 dados"
v = "Â¿La respuesta ofrece argumentos o ejemplos para respaldar su afirmaciÃ³n?"
prompt = template(q, a, v)
prompt
```

```python
%%time
chatgpt(prompt, temp=0.7, max_t=5)["choices"][0]["message"]["content"]
```

```python
test = pd.read_excel("data/test_task_C1.xlsx")
test
```

```python
q, a, label = test.sample(1).iloc[0]["Q A label".split()]
q, a, label
```

```python
%%time
v = "Â¿La respuesta brinda una respuesta directa y relacionada con la pregunta planteada?"
responses = []
for i, ix in enumerate(test.index):
    q, a, label = test.loc[ix]["Q A label".split()]
    a = a.replace("\n", " ")
    a = " ".join(a.split(" ")).strip()
    o = chatgpt(template(q, a, v), temp=0.7, max_t=5)["choices"][0]["message"]["content"]
    o = o.replace("\n", " ").strip()
    responses.append([ix, o])
    print(100*(i+1)/test.shape[0], ix, o, label)
```

```python
test_v1 = test.copy()
test_v1["chatgpt"] = [x[1] for x in responses]
test_v1
```

```python
test_v1.to_excel("chatgpt_responses_test_V1.xlsx")
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(test_v1["label"], test_v1["chatgpt"].apply(lambda x: int("No" in x)), digits=4))
```
