```python
clf = DecisionTreeClassifier(random_state=42, min_impurity_decrease=0.0000, max_depth=5, criterion="gini")
clf.fit(train_sample[new_best_features], train_sample["label"])
```

```python
from matplotlib import pyplot as plt
from sklearn import tree
```

```python
fig, ax = plt.subplots(1, 1, figsize=(265, 85))
tree.plot_tree(clf, feature_names=new_best_features, ax=ax, fontsize=45, class_names=["C0","C1"]);
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(train_sample["label"], clf.predict(train_sample[new_best_features]), digits=4))
```

```python
val_sub_index = val_sample.index
print(classification_report(val_sample.loc[val_sub_index]["label"], clf.predict(val_sample.loc[val_sub_index][new_best_features]), digits=4))
```

```python
sub_index_test = test_sample.index#[test_sample["label_Q"] == 3].index
print(classification_report(test_sample.loc[sub_index_test]["label"], clf.predict(test_sample.loc[sub_index_test][new_best_features]), digits=4))
```
